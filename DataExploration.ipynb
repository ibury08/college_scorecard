```python
#Dropping unnamed columns in dictionary - matching on 'VARIABLE NAME' later.
data_dict = data_dict[~data_dict['VARIABLE NAME'].isnull()]
```

```python
cat_dict={}
cols={}
for row in data_dict.iterrows():
    cols[row[1]['VARIABLE NAME']] = {'name':row[1]['NAME OF DATA ELEMENT'],'notes' : row[1]['NOTES'],'dev-category':row[1][1]}
```

```python
#Dropping data categories we don't want
for k in data:
    for c in data[k].columns:
        try:
            if cols[c]['dev-category'] == 'root':
                data[k].drop(c, axis=1, inplace=True)
        except:
            continue
```

```python
#Taking a look at how often columns are populated by year
populated_cols  = pd.DataFrame()
for year in data.keys():
    populated_cols[year]= data[year].notnull().sum()
populated_cols
```

```python
#Finding columns where we don't have data for each of the last 10 years
bool_cols  = pd.DataFrame()
for year in data.keys():
    bool_cols[year]= data[year].notnull().sum() > 0
bool_cols['Total'] = bool_cols.sum(axis=1)
```
